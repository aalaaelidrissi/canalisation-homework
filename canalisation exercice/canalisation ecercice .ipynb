```python
import numpy as np
D1=float(input("la valeur du diam√®tre 1:"))
V1=float(input("la valeur de la vitesse 1:"))
alpha=float(input("lavaleur de l'angle en degre:"))
V2=1.5*V1
D2=np.sqrt(V1/V2)*D1
L=(-D1/2*np.tan(alpha))*(np.sqrt(V1/V2)-1)
print("la vitesse de section s2 est:",V2,"m/s")
print("la longueur est:",L,"m")
```

```python
from IPython import display
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
Font= ImageFont.truetype('arial',20)
img =Image.open('./image.png')
I1=ImageDraw.Draw(img)
I1.text((135, 145),"D1=200m",fill=(255,0,0) ,font=Font)
I1.text((262, 228),"V1="+str(V1),fill=(255,0,0) ,font=Font)
I1.text((408, 88),"alpha="+str(alpha),fill=(255,0,0) ,font=Font)
I1.text((497, 234),"L="+str(L),fill=(255,0,0) ,font=Font)
I1.text((775, 225),"V2="+str(V2),fill=(255,0,0) ,font=Font)
img.save("./image.png")
display.Image("./image.png")
```
